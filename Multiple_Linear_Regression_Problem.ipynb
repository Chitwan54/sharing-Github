## Linear Regression

```python
# Create a numpy array of 1's for stacking X0
ones = np.ones((X.shape[0],1))

# Stack it with X
X = np.hstack((ones, X))
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
# Generate Hypothesis
def hypothesis(X, theta):
    
    return np.dot(X,theta)

# error function
def total_error(X,y, theta):
    
    m,n = X.shape
    error = 0.0
    
    for i in range(m):
        
        # Call the hypothesis function
        y_ = hypothesis(X[i], theta)
        
        # Calcuate the error function
        error += (y - y_)**2
        
    return error/m

# Gradient
def gradient(X, y, theta):
    
    m,n = X.shape
    grad = np.zeros((n,))
    
    for j in range(n):
        for i in range(m):
            
            # Call the hypothesis
            y_ = hypothesis(X[i], theta)
            
            # Gradient
            grad[j] += (y[i] - y_)*X[i][j]
            
    return grad/m

# Gradient Descent
def gradient_descent(X, y, learning_rate = 0.01, max_iter = 75):
    
    m,n = X.shape
    
    # Randomly initialize thetas
    theta = np.zeros((n,))
    
    # Create an error list
    error_list = []
    
    # Perform the update rule
    for i in range(max_iter):
        
        # Call the gradient
        grad = gradient(X, y, theta)
        
        # Call the error function
        e = total_error(X, y, theta)
        error_list.append(e)
        
        # Perform the update rule
        for j in range(n):
            
            theta[j] = theta[j] - learning_rate* grad[j]
    
    return theta, error_list    
```

```python
(theta, error_list) = gradient_descent(X_train, y_train)
```

## Make Predictions

```python
# Make Predictions on the training data set
def predictions(X,Y, theta):
    y_ = []
    m,n = X.shape
    for i in range(m):
        predictions = hypothesis(X[i], theta)
        y_.append(predictions)
    y_ = np.array(y_)
    return y_

# Calculate the R2 score
def R2_score(y, y_predictions):
    
    num = np.sum((y - y_predictions)**2)
    den = np.sum((y-y.mean())**2)
    
    r2_score = 1 - (num/den)
    return r2_score*100
```

```python
y_train_predictions = predictions(X_train, y_train, theta)
```

```python
R2_score(y_train, y_train_predictions)
```

```python
y_test_predictions = predictions(X_test, y_test, theta)
```

```python
R2_score(y_test, y_test_predictions)
```
